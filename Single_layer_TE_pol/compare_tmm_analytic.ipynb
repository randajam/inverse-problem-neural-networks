# Compare tmm implementation with analitic for one_layer

```python
from physics import *
import numpy as np
import pandas as pd
from itertools import product
```

## Generation input data

```python
h_values = np.linspace(0.1, 5, 50)
k1_values = np.linspace(0.1, 5, 50)
k0, k2 = 1, 1
theta = np.deg2rad(30)
theta_values = np.deg2rad(np.linspace(10, 80, 40))
```

## Compare

```python
r_analitic = []
r_tmm = []
for k1, h, theta in product(k1_values, h_values, theta_values):
    r_analitic.append(r_total([k0, k1, k2], h, theta))
    r_tmm.append(tmm_reflection([k0, k1, k2], [h], theta))

r_analitic = np.array(r_analitic)
r_tmm = np.array(r_tmm)
```

```python
err = np.max(np.abs(r_analitic - r_tmm))
print(f"Max |Δr| = {err:.3e}")
```

Таким образом получаем, что метод tmm написан корректно, поскольку наблюдаемая ошибка в сравнении с ручным выводом формулы находится в районе машинного эпсилон на всеразличных входных данных
